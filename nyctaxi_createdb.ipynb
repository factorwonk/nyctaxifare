## Import in NYC Taxi Trip Data

```python
import pandas as pd
trip_data = pd.read_csv('./nyc_taxi_data/trip_data_4.csv', parse_dates=[' pickup_datetime', ' dropoff_datetime'])
# Remove Whitespace in Column Names
trip_data.columns = [x.strip().replace(' ', '') for x in trip_data.columns]
trip_data.head()
```

```python
trip_data.info()
```

```python
trip_data.tail()
```

## Import NYC Taxi Trip Fare Data

```python
import pandas as pd
fare_data = pd.read_csv('../nyc_taxi_data/trip_fare_4.csv', parse_dates = [' pickup_datetime'])
# Remove whitespace from column names
fare_data.columns = [x.strip().replace(' ', '') for x in fare_data.columns]
fare_data.head()
```

```python
fare_data.info()
```

```python
fare_data.sort_values('pickup_datetime').tail()
```

## Convert to SQLite DataBase

```python
import sqlite3
# Create a sqlite3 connection to nyctaxidb
cnxn = sqlite3.connect('nyctaxidb.sqlite')
```

```python
# This takes a while...
```

```python
# Create trips table
trip_data.to_sql('trips',cnxn)
# Create fares table
fare_data.to_sql('fares',cnxn)
```
